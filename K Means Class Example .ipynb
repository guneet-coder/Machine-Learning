```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('data4.csv')
df
```

```python
def dis(x,y,center):
    return float(math.sqrt((float(x)-float(center[0]))**2+(float(y)-float(center[1]))**2))
```

```python
k = int(input("Enter Number of clusters Required : "))
arr = []
for i in range(k):
    print("For centeroid C{} :".format(i))
    x = input("x = ")
    y = input("y = ")
    arr.append((x,y))
    
#3 1 5 4 1 8 4 sample TC
```

```python
# initial filling of c1,c2,c3
for i in range(k):
    df[i] = [dis(float(df['x'][j]),float(df['y'][j]),arr[i]) for j in range(df.shape[0])]    
df["result"] = df[[0,1,2]].idxmin(axis = 1)
```

```python
new_arr = []
dfx = df.groupby('result').mean()
for x in range(4):
    new_arr = []
    for i in range(dfx.shape[0]):
        new_arr.append((float(dfx['x'][i]),float(dfx['y'][i])))
    for j in range(k):
        df[j] = [dis(float(df['x'][k]),float(df['y'][k]),new_arr[j]) for k in range(df.shape[0])]
    df["result"] = df[[0,1,2]].idxmin(axis = 1)
    dfx = df.groupby('result').mean()
print(new_arr)
```
